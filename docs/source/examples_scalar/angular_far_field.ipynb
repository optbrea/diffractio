# CZT with z=f(x) and r=f($\theta$) observation points

```python
from diffractio import degrees, mm, um
from diffractio import np, plt
from diffractio.diffractio import Diffractio
```

## $z=f(x)$

Sometimes we want to obtain the field at a curved observaton surface. Then we can use the CZT_points function.

```python
x0 = np.linspace(-100*um, 100*um, 1024)
wavelength = .5*um

u0 = Diffractio('scalar', 'source', x=x0, wavelength=wavelength)
u0.plane_wave()

u0.draw('field')
```

```python
x_obs = np.linspace(-10*um, 10*um, 1000)
z_obs = 100*um + np.tan(4*degrees)*x_obs
```

```python
u_far0 = u0.CZT_points(z_obs, x_obs, verbose=False, has_draw=False)

I_far0 = np.abs(u_far0)**2

plt.plot(x_obs, np.angle(u_far0))
plt.title('phase')
plt.xlabel('x (microns)')
```

As the observation plane is inclined, then the phase is not constant.

## $r=f(\theta)$

Also, n diffraction, sometimes we want to obtain the far field  diffraction pattern. Then, we can use the CZT_points CZT_angular function where the input parameters are the radius and the angles of observation

```python
x0 = np.linspace(-50*um, 50*um, 1024*2)
wavelength = .5*um

u0 = Diffractio('scalar', 'source', x=x0, wavelength=wavelength)
u0.plane_wave(A=1, theta=0*degrees)

t = Diffractio('scalar', 'mask', x=x0, wavelength=wavelength, n_background=1)
t.slit(x0=0, size=2.5*um)

u_near=u0*t
```

```python
theta = np.linspace(-60*degrees, 60*degrees, 512)
radius_obs = 20*mm
```

```python
u_far = u_near.CZT_angular(radius_obs=radius_obs, theta=theta, verbose=True, has_draw='all')
```
