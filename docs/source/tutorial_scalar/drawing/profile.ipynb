# Drawing profiles

Here we show how to draw profiles. There are two procedures: profile and profile_angle:

- profile(): point1 to point2, also avaliable npixels.


- profile_angle():  r0 and angle, also availabel npixels.
 

```python
from diffractio import np, plt
from diffractio import mm, um, degrees, nm
from diffractio.diffractio import Diffractio
```

```python
x = np.linspace(-1.0*mm, 1.0*mm, 512)
y = np.linspace(-1.0*mm, 1.0*mm, 512)
wavelength = 550 * nm

t = Diffractio('scalar', 'mask', x=x, y=y, wavelength=wavelength)
t.slit(x0=0, size=150 * um)
u0 = t.RS(z=100 * mm)
u0.draw('intensity')
```

```python
x_coords, y_coords, s_line, vals = u0.profile(kind='intensity', point1 = (-1*mm, 0), point2 = (1*mm, 0),  npixels=800, has_draw=True)
```

```python
x_coords, y_coords, s_line, vals = u0.profile_angle(kind='intensity', r0=(0, 0), angle=45*degrees, length=None, has_draw=1)
```

```python
x_coords, y_coords, s_line, vals = u0.profile_angle(kind='intensity', r0=(0, 0), angle=30*degrees, length=None, npixels=800, has_draw=2)
```
